# Top 10 Major Countries of the world

```python
major_countries = df[df['Country'].isin(['India', 'USA','China','Brazil','France','Russia','Great Britain','Japan','Germany','South Africa']) ].reset_index()
major_countries
```

# **importing datetime**

```python
import datetime as dt
```

**converting object dtype of Date column into datetime dtype**

```python
major_countries["Date"] = pd.to_datetime(major_countries["Date"])
```

**making new columns from Date column : Month , week , day**

```python
major_countries["month"] = major_countries["Date"].dt.month
major_countries["day"] = major_countries["Date"].dt.day
major_countries["week"] = major_countries["Date"].dt.week


major_countries.head()
```

```python
#checking information about dataset
major_countries.info()
```

**Pie chart of major countries Average AQI values over all dates.**

```python
fig = px.pie(major_countries , names = "Country" ,values = "AQI Value", title = "AQI value of major countries", hover_data = ["Status"] )
fig.update_traces(textposition='inside', textinfo='percent+label')
```

**line plot of major countries AQI value**

```python

px.line(major_countries ,x = 'Date', y = 'AQI Value', color = "Country" , title = "AQI values of major countries")

```

**bar plot of major countries AQI value**

```python
px.bar(major_countries ,x = 'month', y = 'AQI Value', color = "Country" , title = "AQI value of major countries according to month" )
```
